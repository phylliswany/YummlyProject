```python
import numpy as np
```

Read the training data set

```python
X = np.load('TrainingDataX.npy')
y = np.load('TrainingDatay.npy')
# print X_train
# print y_train
```

```python
import copy
```

```python
def FlavorClass(z):
    x = []
    for y in z:
        if y <= 17:
            x.append(0)
        elif y >= 83:
            x.append(2)
        else:
            x.append(1)
    # print z[0]
    # print x[0]
    x = np.array(x)
    return x
```

Test the model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import KFold
import matplotlib.pyplot as plt
```

RandomForest for salty (100 is the best)

```python
n_trees = 10
kf = KFold(np.shape(X)[0], n_folds=5)
forest = RandomForestClassifier(n_estimators = n_trees, random_state = 1)
column = 0
accuracy = []
for train_index, test_index in kf:
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index, column], y[test_index, column]
    
    forest.fit(X_train, y_train)
    y_hat = forest.predict(X_test)
    
    accuracy.append([n_trees, float(sum(y_hat == y_test))/len(y_test), float(sum(FlavorClass(y_hat) == FlavorClass(y_test)))/len(y_test)])
    
%matplotlib notebook
plt.plot([x[1] for x in accuracy])
plt.plot([x[2] for x in accuracy])
print np.mean([x[1] for x in accuracy])
print np.mean([x[2] for x in accuracy])
```
